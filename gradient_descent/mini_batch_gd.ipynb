```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import random
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
class MBGDRegressor:
    
    def __init__(self,batch_size,learning_rate=0.01,epochs=225):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        
    def fit(self,X_train,y_train):
       
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])  # initialize the coefficients matrix to 1
        
        for i in range(self.epochs):
            
            for j in range(int(X_train.shape[0]/self.batch_size)):        # no. of batches = no. of rows / batch size
                
                idx = random.sample(range(X_train.shape[0]),self.batch_size)   # take a random sample of rows from the training data of batch size
                
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
               
                intercept_der = -2 * np.mean(y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
mbr = MBGDRegressor(batch_size=int(X_train.shape[0]/25),learning_rate=0.01,epochs=250)
```

```python
mbr.fit(X_train,y_train)
```

```python
y_pred = mbr.predict(X_test)
```

```python
r2_score(y_test,y_pred)*100
```
